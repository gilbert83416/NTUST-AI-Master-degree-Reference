```python
# 視野空間的改動
from tensorflow.keras.applications.inception_v3 import InceptionV3
cnn = InceptionV3(include_top=False, input_shape=(224, 224, 3))
cnn.summary()
```

```python
# stride = 2 -> 一起做poolingz size = 4 避免棋盤效應
from tensorflow.keras.utils import plot_model
plot_model(cnn, to_file="inception.png", show_shapes=True)
```

```python
print(len(cnn.layers))
```

```python
# 1 * 1
w = cnn.get_layer("conv2d_5").get_weights()
w[0].shape
```

```python
# # 1 * 1
w = cnn.get_layer("conv2d_11").get_weights()
w[0].shape
```

```python
# # 5 * 5 = 1 * 1 -> 5 * 5
w = cnn.get_layer("conv2d_6").get_weights()
print(w[0].shape)
w = cnn.get_layer("conv2d_7").get_weights()
print(w[0].shape)
```

```python
# 5 * 5 = 1 * 1 -> 3 * 3 -> 3 * 3
w = cnn.get_layer("conv2d_8").get_weights()
print(w[0].shape)
w = cnn.get_layer("conv2d_9").get_weights()
print(w[0].shape)
w = cnn.get_layer("conv2d_10").get_weights()
print(w[0].shape)
```
